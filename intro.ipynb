```python
img = cv.imread("Lenna.png")
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)
print(img_gray.shape)
plt.imshow(img_gray)
plt.show()
```

```python
img_100 = img_gray
img_100[img_100>100] = 255
img_100[img_100<100] = 0
plt.imshow(img_100)
plt.show()
```
